```python
import pandas as pd
import numpy as np

# Dataframe original
df_original = pd.read_csv('/kaggle/input/airbnbopendata/Airbnb_Open_Data.csv', low_memory=False)
# Vamos fazer uma c√≥pia para trabalharmos
df = df_original.copy()
df.head(10)
```

Drop the last column as it doesnt contain anything.

```python
df.drop('license', axis=1,inplace = True)
```

```python
df.info()
```

Treat missing values. I will assume that column 21 is important to keep. So I will subset the dataset based on nonnull values from this column

```python
df = df[df['review rate number'].notnull()]

df.info()
```

Change data type

```python

df['Construction year'] = pd.to_datetime(df['Construction year'], format='%Y', errors='coerce')

df.info()
```

Clean up the price columns which contain strings

```python
# Check the current datatype 
df['price'].dtypes
```

```python
# Check the current datatype 
df['service fee'].dtypes
```

```python
# This function will remove the '$' and ',' from each individual element of the column and return float(x) or np.nan (for nan values)

def clean_column(x):
    try:
        x = x.replace('$', '')
        x = x.replace(',', '')
        return float(x)
    except:
        return np.nan

df['price'] = df['price'].apply(clean_column)
```

```python
df['service fee'] = df['service fee'].apply(clean_column)
```

```python
df.info()
```
