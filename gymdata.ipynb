### Are there gender differences in workouts?



Compare Avg_BPM and Max_BPM between men and women

```python
sns.boxplot(data=df, x='Gender', y='Max_BPM')
```

```python
sns.boxplot(data=df, x='Gender', y='Avg_BPM')
```

```python
df_melted = pd.melt(df, id_vars=['Gender'], value_vars=['Max_BPM', 'Avg_BPM'],
                    var_name='Heart_Rate_Type', value_name='BPM')

df_melted
sns.boxplot(data=df_melted, x='Gender', y='BPM')
```

TTest to check for differences

```python
from scipy.stats import ttest_ind

men_avg_bpm = df[df['Gender'] == 'Male']['Avg_BPM']
women_avg_bpm = df[df['Gender']=='Female']['Avg_BPM']

ttest, pvalue=ttest_ind(men_avg_bpm, women_avg_bpm)
print(f"Avg_BPM t-test: t-statistic={ttest}, p-value={pvalue}")
```

No statistical significance in Avg_BPM between men and women. Also demonstrated below.

```python
df[df['Gender'] == 'Male']['Avg_BPM'].describe()
```

```python
df[df['Gender'] == 'Female']['Avg_BPM'].describe()
```
